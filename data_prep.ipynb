```python
import pandas as pd
import requests
import urllib.parse
import time
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_xml("/Users/jmawi/Downloads/food_premises.xml")

#df = pd.read_xml("https://ratings.food.gov.uk/api/open-data-files/FHRS807en-GB.xml")
```

```python
supermarkets_df = df[df["BusinessType"] == "Retailers - supermarkets/hypermarkets"].reset_index(drop=True)
```

```python
supermarkets_df.head(2)
```

```python
supermarkets_df.shape
```

```python
supermarkets_df["lookup_address"] = supermarkets_df["AddressLine3"] + ", " + supermarkets_df["AddressLine4"]
```

```python
supermarkets_df.head(2)
```

```python
def get_lat_long_from_address(address_list):
    lats = []
    longs = []
    for address in address_list:
        url = 'https://nominatim.openstreetmap.org/search?q=' + urllib.parse.quote(address) +'&format=json'
        try:
            response = requests.get(url).json()
            lats.append(response[0]["lat"])
            longs.append(response[0]["lon"])
            time.sleep(1)
        except:
            print("Error fetching {}".format(address))
            lats.append("missing")
            longs.append("missing")
            
        
    return lats, longs
```

```python
list(supermarkets_df["lookup_address"].iloc[:5])
```

```python
the_lats, the_longs = get_lat_long_from_address(list(supermarkets_df["lookup_address"]))
```

```python
supermarkets_df["latitude"] = the_lats
supermarkets_df["longitude"] = the_longs
```

```python
supermarkets_df.head(2)
```

```python
supermarkets_df2 = supermarkets_df[supermarkets_df["latitude"]!="missing"].reset_index(drop=True)
```

```python
supermarkets_df2.head(2)
```

```python
supermarkets_df2.to_csv("supermarkets_with_lat_long.csv")
```
